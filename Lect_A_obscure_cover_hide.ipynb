```python
from datascience import *
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

# Data Cleaning

```python
gss = pd.read_csv('socio.csv')[:-1]
gss
```

```python
# check the data types of each variable
gss.dtypes
```

```python
# numerical variables are data type 'object', not int & change 'not applicable' to -1
# we need to convert it to do aggregate functions

gss['YEAR'] = gss['YEAR'].fillna(-1)
gss['REALINC ($)'] = pd.to_numeric(gss['REALINC ($)'], errors = 'coerce').fillna(np.nan)
gss['TVHOURS'] = pd.to_numeric(gss['TVHOURS'], errors = 'coerce').fillna(np.nan)
gss['EDUC'] = pd.to_numeric(gss['EDUC'], errors = 'coerce').fillna(np.nan)
gss['CHILDS'] = pd.to_numeric(gss['CHILDS'], errors = 'coerce').fillna(np.nan)
gss['SIBS'] = pd.to_numeric(gss['SIBS'], errors = 'coerce').fillna(np.nan)
gss['HRS1'] = pd.to_numeric(gss['HRS1'], errors = 'coerce').fillna(np.nan)
gss['REALRINC'] = pd.to_numeric(gss['REALRINC'], errors = 'coerce').fillna(np.nan)
gss.drop(64814, inplace = True)

gss.dtypes
gss
```

```python
gss.to_csv('cleaned_data.csv')
```

```python
def lineplot(x, y, data):
    sns.lineplot(x = x, y = y, data = data)
    plt.xlim(1970, 2018)
```

```python
def distplot(column, data, color):
    plt.figure(figsize=(8,4))
    sns.distplot(data[column], color = color)
```

```python
def mode_distplot(column, data, color): 
    plt.figure(figsize=(8,4))
    sns.distplot(data[column], color = color, kde= False, rug = False)
```

```python
def marker(value):
    plt.axvline(value, color = 'black', linewidth = 1.3)
```

```python
#marker in diff color (for comparisons)
def second_marker(value):
    plt.axvline(value, color = 'yellow', linewidth = 1.3)
```

```python
def mean(data, column):
    return np.mean(data[column])
```

```python
def show_greater_than(data, column, value):
    return data[data[column] > value]
```

```python
def show_less_than(data, column, value):
    return data[data[column] < value]
```

```python
def show_greater_equal(data, column, value):
    return data[data[column] >= value]
```
